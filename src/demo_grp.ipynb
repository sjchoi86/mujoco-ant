# Gaussian Random Path

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from lgrp_class import lgrp_class
print ("Packages loaded.")
```

### Leveraged Gaussian Random Path

```python
# Set anchor points and time indices
tMin,tMax = 0,3
dim = 4
nData = 10 # 2: GP Prior / 10: GP Posterior-ish
dT = 0.01
nTest = (int)((tMax-tMin)/dT) 
# Set data 
tData = np.linspace(start=tMin,stop=tMax,num=nData).reshape((-1,1))
xData = np.random.rand(nData,dim) # Random positions 
lData = np.ones(shape=(nData,1))
lData[1:nData-1] = 0.5
tTest = np.linspace(start=tMin,stop=tMax,num=nTest).reshape((-1,1))
lTest = np.ones(shape=(nTest,1))

# Leveraged GRP
hyp = {'gain':1/4,'len':1/3,'noise':1e-8} 
G = lgrp_class(_name='LGRP',_tData=tData,_xData=xData,_lData=lData,
               _tTest=tTest,_lTest=lTest,_hyp=hyp)

# Sampled paths 
sampledPaths = G.sample_paths(_nPath=3) 

# Plot mean, var, and sampled paths
G.plot_all(_nPath=20,_figsize=(15,6),_titleStr='Leveraged GPR')
```
